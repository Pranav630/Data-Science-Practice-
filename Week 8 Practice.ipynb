```python
### test and train split
### training set of about 80% of your data and then after you've used
### 80% to train, you will test the model that you just created 
### and see what the accuracy is 
```

```python
X=X.drop('MortYN', axis='columns')

## please look into how to use encoding for machine learning - this takes non numerical
## (but useful) data and converts it into new columns-usualy with 0 and 1 values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.20,
                                                   random_state=2)
```

```python
print(X_train.shape, y_train.shape)
```

```python
## look into the various solvers that can 
## be used in logistic regression

logmodel=LogisticRegression(solver='newton-cg', max_iter=10000)
```

```python
## look into fitting a model 
logmodel.fit(X_train, y_train)
```

```python
## look into prediction - predict
predictions=logmodel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

# Let's talk in probabilities

```python

#Creating a sub dataframe that has customers with income greater than 110

over_110=file[file['Income']>=110]


# getting the size or couont of the dataframe\
# this is using the built in len function len() which will give you the number of data rows in a df
over_length=len(over_110)


#now counting the number of customers in our over 110 df who have a personal loan
# this is using the built in function value_counts
over_110_yes=over_110['Personal Loan'].value_counts()[1]


#basic probability: the situation you are curious about/the total sample size
# What is the probability of a a customer having a personal loan if they have an income greater than 110
# (in our dataset )

#over_110_yes_prob=(over_110_yes/over_length)*100
over_110_yes_prob=round((over_110_yes/over_length)*100,2)
over_110_yes_prob


```

#### The probability of a a customer having a personal loan if they have an income greater than 110 is 38.3%

```python

#Creating a sub dataframe that has customers with income less than 110

under_110=file[file['Income']<110]

# getting the size or couont of the dataframe
# this is using the built in len function len() which will give you the number of data rows in a df
under_length=len(under_110)


#now counting the number of customers in our under 110 df who have a personal loan
# this is using the built in function value_counts
under_110_yes=under_110['Personal Loan'].value_counts()[1]

#basic probability: the situation you are curious about/the total sample size
# What is the probability of a a customer having a personal loan if they have an income less than 110
# (in our dataset )

#under_110_yes_prob=(under_110_yes/under_length)*100
under_110_yes_prob=round((under_110_yes/under_length)*100,2)
under_110_yes_prob
```

#### The probability of a a customer having a personal loan if they have an income less than 110 is 1.68%
