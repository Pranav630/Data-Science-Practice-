# Filter application for truth

```python
ages = [5, 12, 17, 18, 24, 32, 45, 53, 64,70]

def myFunction(x):
  if x < 18:
    return False
  else:
    return True

adults = filter(myFunction, ages)

for x in adults:
  print(x)
```

#Problem:
    
#### Start at step 0 on a staircase.  Roll a dice, if you roll 1 or 2 go down one step (unless already at step zero).  
#### If you roll a 3-5, go up one step.  If you roll a 6, reroll the dice and whatever number you roll, go up that many
#### steps.  Each time you roll, there is 0.1% chance that you fall down all the steps and have to restart at step 0.  
#### you will roll the dice 100 times.


```python
import random
```

```python
all_walks = []


for i in range(10):
    random_walk=[0]
    for x in range(100):
        step=random_walk[-1]
        dice = np.random.randint(1,7)
        #print(x)
        if dice <=2:
            step=max(0, step-1)
        elif dice<=5:
            step=step+1
        else:
            step = step + np.random.randint(1,7)
        #print(step)
        
        #implement clumsiness
        rand_prob=np.random.rand()
        if rand_prob<=0.001:
            step =0
        
        #using the build in append function
        random_walk.append(step)
        
    all_walks.append(random_walk)
    

        
    
```

```python
for i in range(1) :
    random_walk = [0]
    for x in range(100) :
        #our next step is applied to our last location
        step = random_walk[-1]
        #print(random_walk)
        #rolls 1-6
        dice = np.random.randint(1,7)
        #if you roll 1 or 2 go down a step
        if dice <= 2:
            #the lowest step is zero
            step = max(0, step - 1)
        #if you roll 3-5 you go up one step
        elif dice <= 5:
            step = step + 1
            
        else:
            #rolling a 6 is a reroll with that number applied
            step = step + np.random.randint(1,7)
        print(step, dice)

        # Implement clumsiness
        rand_prob=np.random.rand()
        #print(rand_prob)
        if rand_prob<=0.001 :
            
            
            step = 0

        random_walk.append(step)
        
    
    # Append random_walk to all_walks
    all_walks.append(random_walk)
```
