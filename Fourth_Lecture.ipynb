## Factorial of a number is the product of all the integers from 1
## to that number. For example, the factorial of 6 (denoted as 6!) is 1* 2* 3* 4* 5* 6 = 720.

```python
x=5
```

```python
def factorial_practice(x):
    """This is a recursive function
    to find the factorial of an integer"""

    if x == 1:
        return 1
    else:
        return (x * factorial_practice(x-1))


num=x
print("The factorial of", num, "is", factorial_practice(num))
```

```python
factorial_practice(5)
```

```python
factorial_practice(x)
```
